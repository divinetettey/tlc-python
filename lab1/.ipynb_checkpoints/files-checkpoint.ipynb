```python
with open("file.txt", 'a') as f: 
    line = ''
    while (line != 'end'): 
        line  = input("hi, input a line of text")
        f.write(line+"\n")
    
```

```python
try:
    f = open("file.txt",'r');
    f2 = open("file_2.txt",'a')
    lines = f.readlines();
    
    for line in lines:
        print("Reading line: "+line)
        f2.write('\n'+line)
    f2.close()
    f.close()
except:
    print('An Error occured')
        
```

```python
try:
    f1 = open('file.txt','r');
    f2 = open('file_2.txt','r');
    f3 = open('new_file.txt','r');
    
    newFile = open('brand_new.txt','a')
    
    for line in f1: 
        newFile.write(line)
    for line in f2: 
        newFile.write(line)
    for line in f3:
        newFile.write(line)
        
    f1.close()
    f2.close()
    f3.close()
    newFile.close();
except:
    print('Error occured')
```

```python
with open("file_2.txt","r") as f: 
    lines = f.readlines()
    lines.reverse();
    newFile = open('new_file.txt','w')
    
    for line in lines:
        print("Writing line: "+line)
        newFile.write('\n'+line)
    newFile.close()
    
```
